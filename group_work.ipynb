```python
data = pd.read_csv('FremontBridge.csv')
```

```python
renamed_data = data.rename(columns = {'Fremont Bridge East Sidewalk':'East','Fremont Bridge West Sidewalk':'West'})
renamed_data
```

```python
renamed_data['Total'] = renamed_data.eval('West + East')
```

```python
renamed_data.dropna().describe()
```

```python
%matplotlib inline
import seaborn; seaborn.set()
```

```python
renamed_data.plot()
plt.ylabel('hourly Bicycle Count');
```

```python
renamed_data.index = pd.to_datetime(renamed_data['Date'])
```

```python
weekly = renamed_data.resample('W').sum()
weekly.plot(style = [':','--','-'])
plt.ylabel('Weeliy bicycle count')
```

```python
daily = renamed_data.resample('D').sum()
daily.rolling(30,center = True).sum().plot(style = [':','--','-'])
plt.ylabel('mean hourly count')
```

```python
daily.rolling(50,center = True,win_type='gaussian').sum(std=10).plot(style = [':','--','-']);
```

```python
by_time = renamed_data.groupby(renamed_data.index.time).mean()
hourly_ticks = 4*60*60*np.arange(6)
by_time.plot(xticks = hourly_ticks, style = [':','--','-']);
```

```python
by_weekday = renamed_data.groupby(renamed_data.index.dayofweek).mean()
by_weekday.index = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
by_weekday.plot(style = [':','--','-']);
```

```python
weekend = np.where(renamed_data.index.weekday < 5, 'Weekday', 'Weekend')
by_time = renamed_data.groupby([weekend, renamed_data.index.time]).mean()
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,5))
by_time.loc['Weekday'].plot(ax=ax[0],title='Weekdays',xticks = hourly_ticks, style = [':','--','-'])
by_time.loc['Weekend'].plot(ax=ax[1],title='Weekends',xticks = hourly_ticks, style = [':','--','-']);
#ixはもう古い。locかilocを使う。
```
